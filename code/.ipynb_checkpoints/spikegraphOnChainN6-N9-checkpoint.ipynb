```python
N_L = np.arange(6, 10)
```

```python
for N in N_L:
    cfg = CFG(N=N)
    W = np.zeros((cfg.N, cfg.N))

    for i in range(1, cfg.N):
        W[i, i - 1] = 3.5

    parents = [[]] + [[i - 1] for i in range(1, cfg.N)]

    if __name__ == "__main__":
        main(cfg, W, parents)
```
