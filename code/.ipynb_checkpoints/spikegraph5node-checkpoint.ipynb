```python
from spikegraph import *
```

### Chain 0 -> 1 -> 2 -> 3 -> 4

```python
cfg = CFG(N=5)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[4, 3] = 3.5
parents = [[], [0], [1], [2], [3]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```

### 5-cycle 0 -> 1 -> 2 -> 3 -> 4 -> 0

```python
cfg = CFG(N=5, K=10000, T=60)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[4, 3] = W[0, 4] = 3.5
parents = [[4], [0], [1], [2], [3]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```

### 5-cycle 0 -> 1 -> 2 -> 3 -> 4 -> 0 and a chord 0 -> 2

```python
cfg = CFG(N=5, K=10000, T=60, nll_coeff=10)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[4, 3] = W[0, 4] = W[2, 0] = 3.5
parents = [[4], [0], [0, 1], [2], [3]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```

### 5-cycle 0 -> 1 -> 2 -> 3 -> 4 -> 0 and two chords 0 -> 2, 3 -> 1

```python
cfg = CFG(N=5, K=10000, T=60, nll_coeff=10)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[4, 3] = W[0, 4] = W[2, 0] = W[1, 3] = 3.5
parents = [[4], [0, 3], [0, 1], [2], [3]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```

### Combo: Chain 0->1->2, Inward fork 4->1<-3, Cycle 1->2->3->1

```python
cfg = CFG(N=5, K=10000, T=60)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[1, 3] = W[1, 4] = W[1, 3] = 3.5
parents = [[], [0, 3, 4], [1], [2], []]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```

### Two connected cycles  0 -> 1 -> 2 -> 0, 1 -> 3 -> 4 -> 1

```python
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[0, 2] = 3.5
W[3, 1] = W[4, 3] = W[1, 4] = 3.5
parents = [[2], [0, 4], [1], [1], [3]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

```python
plt.imshow(W)
```
