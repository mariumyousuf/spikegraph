### Chain 

```python
cfg = CFG(N=10)
W = np.zeros((cfg.N, cfg.N))

for i in range(1, cfg.N):
    W[i, i - 1] = 3.5
    
parents = [[]] + [[i - 1] for i in range(1, cfg.N)]

if __name__ == "__main__":
    main(cfg, W, parents)
```

### Cycle

```python
cfg = CFG(N=10, T=60, K=10000)
W = np.zeros((cfg.N, cfg.N))

# Create cycle: each node i has a parent (i - 1) % N
for i in range(cfg.N):
    W[i, (i - 1) % cfg.N] = 3.5

# Define parents: each node i has a parent (i - 1) % N
parents = [[(i - 1) % cfg.N] for i in range(cfg.N)]

if __name__ == "__main__":
    main(cfg, W, parents)
```
