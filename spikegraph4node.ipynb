### Chain 0 -> 1 -> 2 -> 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = 3.5
parents = [[], [0], [1], [2]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Inward fork 0 -> 1 <- 2, 1 <- 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[1, 2] = W[1, 3] = 3.5
parents = [[], [0, 2, 3], [], []]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Single edge 0 -> 1 2 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = 3.5
parents = [[], [0], [], []]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Outward fork 0 <- 1 -> 2, 1 -> 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
W[0, 1] = W[2, 1] = W[3, 1] = 3.5
parents = [[1], [], [1], [1]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Cycle 0 -> 1 -> 2 -> 3 -> 0

```python
cfg = CFG(N=4, K=10000, T=60)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[0, 3] = 3.5
parents = [[3], [0], [1], [2]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Empty 0 1 2 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
parents = [[], [], [], []]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### DAG 0 -> 1 -> 2 -> 3, 0 -> 3

```python
cfg = CFG(N=4)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[3, 0] = W[2, 1] = W[3, 2] = 3.5 
# W[0, 1] = W[1, 2] = W[0, 2] = 3.5
parents = [[], [0], [1], [0, 2]]
if __name__ == "__main__":
    main(cfg, W, parents)
```

### Cycle and a chord 0 -> 1 -> 2 -> 3 -> 0, 0 -> 2

```python
# N: int = 3
# T: int = 20 #60
# K: int = 1000 #10000
# seed: int = 123
# lr_tab: float = 1e-1
# iters_tab: int = 500
# lr_nn: float = 3e-4
# iters_nn: int = 1500
# hidden: int = 64
# batch_m: int = 256
# nll_coeff: float = 1.0
# device: str = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
cfg = CFG(N=4, K=10000, T=60, nll_coeff=10.0)
W = np.zeros((cfg.N, cfg.N))
W[1, 0] = W[2, 1] = W[3, 2] = W[0, 3] = W[2, 0] = 3.5
parents = [[3], [0], [0, 1], [2]]
if __name__ == "__main__":
    main(cfg, W, parents)
```
